```python
from pyspark import SparkConf,SparkContext
```

```python
conf=SparkConf().setAppName("demo").setMaster("local")
```

```python
sc=SparkContext(conf=conf)
```

```python
#Create RDD Transform
rdd1=sc.parallelize(['a1','a2'])
```

```python
# @title Default title text
#Action on RDD
rdd1.collect()
```
